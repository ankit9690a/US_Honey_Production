### Q2) Which are the top 5 Honey producing states in the US?

```python
df2=df.groupby('state').sum().sort_values(by='production',ascending=False).head()
df2['production']
```

```python
#before visualizing in graphs we need to reset the indexing of new data i.e. df2
df2=df.groupby('state').sum().reset_index().sort_values(by='production',ascending=False).head()
df2['production']
```

```python
sns.barplot(x=df2['state'],y=df2['production'])
plt.show()
```

### Q3) What is the Change in mean Average price of Honey from 1995 to 2021?

```python
df3=df.groupby('year').mean('average_price').reset_index()
df3
```

```python
74.840909-3.334250
```
